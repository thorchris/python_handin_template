```python
##Imports
import bs4
import requests
import re
import mysql.connector as mysql
```

```python
##Exercise 1
##Use BeautifulSoup, to select the name and price for the laptops (hint: look for class="product-price" to begin with).
r = requests.get('https://www.komplett.dk/category/21661/gaming/gaming-pc/baerbar')
r.raise_for_status()
soup = bs4.BeautifulSoup(r.text, 'html.parser')

all_laptops = {}

name = soup.select('div[class=text-content] > h2')
price = soup.select('span[class=product-price-now]')

for idx, val in enumerate(name):
    all_laptops[name[idx].getText()] = price[idx].getText()
    
print(all_laptops)
```

```python
##Exercise 1 
##Create a mysql table called laptops, with columns id, name, price.

#Done using following commands:
#docker ps to see the runningn contianers
#docker exec -it mysql_notebooks bash to get into the container
#mysql -u root -proot to get into sql prompt
#show databases; use db; show tables; to see databases and to see tables inside the db database.
#CREATE TABLE IF NOT EXISTS laptops (id INT AUTO_INCREMENT PRIMARY KEY, name VARCHAR(255), price VARCHAR(255))
```

```python
##Exercise 1
##Insert the laptop data into the laptops table, using a mysql connector and cursor.

## connecting to the database using 'connect()' method
cnx = mysql.connect(host = "db", user = "root", passwd = "root", db = "db")

cursor = cnx.cursor(prepared=True)

# Insert a laptop
sql = "INSERT INTO laptops (name, price) VALUES (%s %s)" 

values = (all_laptops.keys(), all_laptops.values())

cursor.execute(sql, values)
    
cursor.close()
cnx.close()
```

```python
##Exercise 3 
##Select the data from the endpoints, using a request object.
r2 = requests.get('http://localhost:5000/api/laptops/all')
r2.raise_for_status()
soup2 = bs4.BeautifulSoup(r2.text, 'html.parser')

names_reg = re.compile(r'"name":.([\w\ ]+)')
prices_reg = re.compile(r'"price":.([\w\.]+)')

all_reg_names = names_reg.findall(str(soup2))
all_reg_prices = prices_reg.findall(str(soup2))

print('All matches: {}'.format(all_reg_names))
print('All matches: {}'.format(all_reg_prices))
```

```python
##Make a sorted bar chart of the laptops, where x-axis is names, and y-axis is price.
import matplotlib.pyplot as plt
%matplotlib inline

sorted_dict = {k: v for k, v in sorted(all_laptops.items(), key=lambda item: item[1])}

x = list(sorted_dict.keys())
y = list(sorted_dict.values())

plt.figure(figsize=(20,10))
plt.xticks(rotation=90)

plt.title("Laptops", fontsize=40)
plt.xlabel("Laptop names", fontsize=25)
plt.ylabel("Price", fontsize=25)

plt.bar(x,y, color=['black', 'red', 'green', 'blue', 'cyan'])

```
