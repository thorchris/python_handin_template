```python
#imports
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from itertools import cycle
from sklearn.cluster import MeanShift, estimate_bandwidth
from sklearn.datasets import make_blobs
```

```python
#load 'data/iris_data.csv' into a dataframe (use decimal=',') 
iris_data = pd.read_csv('../data/iris_data.csv', decimal=',')

#Remove the 2 Petal columns. Now we are left with a 2D feature space
iris_data.drop(['Petal length','Petal width'],'columns',inplace=True)

#get unique labels (Species column)
labels = np.unique(iris_data['Species'])
print("Unique labels: ", labels)

#plot with a scatter plot each iris flower sample colored by label (3 different colors)
x = iris_data['Sepal length']
ys = iris_data['Sepal width']

for spec in labels:
    plt.scatter(x, ys, s = (iris_data['Species'] == spec)*30)
```

```python
##use: MeanShift and estimate_bandwidth from sklearn.cluster to first estimate bandwidth and then get the clusters
#(HINT: estimate_bandwidth() takes an argument: quantile set it to 0.2 for best result)
#print out labels, cluster centers and number of clusters (as returned from the MeanShift function)

#Find esimated bandwidth
iris_data.drop(['Species'],'columns',inplace=True)
eb = estimate_bandwidth(iris_data, quantile=0.2)
print("Estimated bandwidth: ", eb)

#Find esimated clusters
analyzer = MeanShift(bandwidth=eb) 
analyzer.fit(iris_data)
labels = analyzer.labels_
cluster_centers =  analyzer.cluster_centers_
n_clusters = len(np.unique(labels))

print('Estimated clusters: {} '.format(np.unique(labels)))
print('Clusters centers: \n', cluster_centers )
```

```python
#create a new scatter plot where each flower is colored according to cluster labe
#add a dot for the cluster centers
data_2d, _  = make_blobs(n_samples=len(iris_data), centers=cluster_centers, cluster_std=0.1)

analyzer.fit(data_2d)
labels = analyzer.labels_
cluster_centers =  analyzer.cluster_centers_
n_clusters = len(np.unique(labels))

fig = plt.figure()
ax = fig.add_subplot(111)

colors = cycle('bgrcmy')
for k, col in zip(range(n_clusters), colors):
    my_members = (labels == k)
    cluster_center = cluster_centers[k]
    
    x, y = data_2d[my_members,0], data_2d[my_members,1]
    ax.scatter(x, y, c=col, linewidth=0.2)
    ax.scatter(cluster_center[0], cluster_center[1], c='k', s=50, linewidth=0.2)
    
plt.title('Estimated number of clusters: {}'.format(n_clusters))
plt.show()
```
